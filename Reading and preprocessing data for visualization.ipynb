```python
#Data cleaning

#Let's rename the columns so that they make sense
df_covid_19.rename(columns={'Province/State':'State','Country/Region':'Country'}, inplace=True)
df_covid_19.columns
```

```python
#add a 'Total' column that sums up the total people by state over the entire period 2020-01-2 to 2020-04-11 , as follows:
df_covid_19['Total'] = df_covid_19.sum(axis=1)
```

```python
#We have a problem here, some of cells related to state are nan which makes it meaningless while visualization, so
#To remove NaN and replace it with sth more reasonable:
df_covid_19= df_covid_19.apply(lambda x: x.fillna("-")) 
df_covid_19.head()
```

```python
#Given to we want to visualize based on states and countries, we need a column describing both state and country. 
#Before that, we need to change the type of the columns to str so that we are able to join them
df_covid_19["Country"]= df_covid_19["Country"].astype(str)
df_covid_19["State"]= df_covid_19["State"].astype(str)
```

```python
#Join two columns
df_covid_19['State_Country'] = df_covid_19[["State", "Country"]].agg('-'.join, axis=1)
```

```python
#to search for data on a country, we need to know the corressponding index value. 
#This can be fixed very easily by setting the 'Country' column as the index using set_index() method
df_covid_19.set_index('State_Country', inplace=True)
# tip: The opposite of set is reset. So to reset the index, we can use df_can.reset_index()
```
